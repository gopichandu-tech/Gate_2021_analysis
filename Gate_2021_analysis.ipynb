```python
Gate_score.rename(columns={'Paper Code':'Paper_Code'},inplace=True)
```

```python
def overtake(Paper_Code):
    if Paper_Code in ['AE','AG','AR']:
        return 'A family_Engineering'
    elif Paper_Code in ['BM','BT']:
        return 'B family_Engineering'
    elif Paper_Code in['CE*','CH','CS*','CY']:
        return 'C family_Engineering'
    elif Paper_Code in ['EC','EE','ES','EY']:
        return 'E family_Engineering'
    elif Paper_Code in ['GG (Geology)','GG (Geophysics)']:
        return 'G family_Engineering'
    elif Paper_Code in ['IN']:
        return 'I family_Engineering'
    elif Paper_Code in ['MA','ME*']:
        return 'M family_Engineering'
    else:
        return 'another family_Engineering'
Gate_score['Engineering_Family'] = Gate_score.apply(lambda x:overtake(x['Paper_Code']),axis=1)
    
```

Traning and Testing the data

```python
X = pd.DataFrame(Gate_score,columns=['Qualified\nCandidates','Non-Qualified\nCandidates'])
```

```python
y = Gate_score['Candidates\nAppeared']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

Standardizing data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression() 
lin_model.fit(X_train, y_train)
```

```python
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

# on testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))

print("The model performance for testing set")
print('RMSE is {}'.format(rmse))

from sklearn.metrics import r2_score
print(r2_score(y_test,y_test_predict))
```

```python
plt.scatter(y_test, y_test_predict)
plt.plot([min(y_test_predict),max(y_test_predict)],[min(y_test_predict),max(y_test_predict)], color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
